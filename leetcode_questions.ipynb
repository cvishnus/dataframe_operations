
when you do this - 

if key in my_dict:

what it does - 

1. Compute hash(key) → O(1)

2. Jump to that memory slot directly → O(1)

3. Check if the stored key matches (in case of hash collision)

That’s it — no need to scan through every entry like a list.

When you do:

my_dict["apple"] = 5


Python performs roughly these steps:

Compute hash("apple") → e.g. -8558595301535314666

Convert that hash to a valid index:

index = hash("apple") % table_size


Store "apple": 5 in that slot.

Later, when you do:

if "apple" in my_dict:


Python:

Computes the same hash again.

Jumps straight to that slot (index).

Checks if the key there is "apple".

If yes → found it immediately.

If no → check nearby slot (collision handling).

All that happens in constant time, O(1) (on average).

```python
# https://leetcode.com/problems/add-two-numbers/

class Solution(object):
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: Optional[ListNode]
        :type l2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        num1=0
        num2=0
        for i in range(len(l1)):
            num1 +=10**(len(l1)-i-1)*l1[i]
        
        for i in range(len(l2)):
            num2 +=10**(len(l2)-i-1)*l2[i]

        final_num = num1+num2
        print(num1,num2)
        list_final_num = [int(i) for i in str(final_num)]
        return list_final_num[::-1]
    # list(final_num)[::-1]
    # 
```
